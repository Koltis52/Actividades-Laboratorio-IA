```python
import pandas as pd
import numpy as np
data = pd.read_csv('Student_Marks.csv')
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
plt.plot(data['time_study'], data['Marks'], "b.")
plt.xlabel("Tiempo de estudio")
plt.ylabel("Notas")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(data['time_study'].values.reshape(-1, 1), data['Marks'].values)
```

```python
lin_reg.intercept_
```

```python
lin_reg.coef_
```

```python
X_min_max = np.array([[data["time_study"].min()], [data["time_study"].max()]])
y_train_pred = lin_reg.predict(X_min_max)
```

```python
plt.plot(X_min_max, y_train_pred, "g-")
plt.plot(data['time_study'], data['Marks'], "b.")
plt.xlabel("Tiempo de estudio")
plt.ylabel("Notas")
plt.show()
```

```python
x_new = np.array([[2]])

nota = lin_reg.predict(x_new) 

print("La nota para ese tiempo de estudio sería: ", int(nota[0]))
```

```python
plt.plot(data['time_study'], data['Marks'], "b.")
plt.plot(X_min_max, y_train_pred, "g-")
plt.plot(x_new, nota, "rx")
plt.xlabel("Tiempo de estudio")
plt.ylabel("Notas")
plt.show()
```
